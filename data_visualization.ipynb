# Examples of data visualization in python

### The examples include:

1. Correlation matrix plot 
2. Histograms
3. Heatmap

```python
#From https://machinelearningmastery.com/visualize-machine-learning-data-python-pandas/
# Correction Matrix Plot
import matplotlib.pyplot as plt
import pandas
import numpy
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pandas.read_csv(url, names=names)
correlations = data.corr()
# plot correlation matrix

fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)

#following code is supposed to save the plot but doesn't actually work
plt.savefig('destination_path.eps', format='eps', dpi=300)
plt.grid(None)
plt.show()
```

```python
#Matrix plot for a 2D 3*3 system

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#amlp10 = np.array([(0.338, 0.279, 0.381),(0.51, 0.167, 0.322),(0.155, 0.478, 0.365)])
#amlp25 = np.array([(0.362, 0.244, 0.392), (0.536, 0.172, 0.29), (0.344, 0.363, 0.291)])
amlp40 = np.array([(0.314, 0.268, 0.416), (0.452, 0.191, 0.356), (0.441, 0.289, 0.268)])

#The next two lines are important to initiate the plot
fig = plt.figure()
ax = fig.add_subplot(111)

#Draw the actual plot and colourbar
cax = ax.imshow(amlp40, cmap='hot_r', vmin=0.15, vmax=0.55)
clb = fig.colorbar(cax)
clb.set_label('Fractional Interactions', fontsize='large') #control size of cbar font

#Put a frame aroun the plot and define its thickness
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams["axes.linewidth"] = 1

#Define the X and Y axes
xnames=['','AMLP','','POPG','','POPE','','','']
yname=['','POPE','','POPG','','AMLP','']
ax.set_xticklabels(xnames, weight='bold')
ax.set_yticklabels(yname, weight='bold')

#Annotate each of the matrix values
ax.text(0,0,0.314,ha='center',va='center')
ax.text(1,0,0.268,ha='center',va='center')
ax.text(2,0,0.416,ha='center',va='center')
ax.text(0,1,0.452,ha='center',va='center',color='w')
ax.text(1,1,0.191,ha='center',va='center')
ax.text(2,1,0.356,ha='center',va='center')
ax.text(0,2,0.441,ha='center',va='center', color='w')
ax.text(1,2,0.289,ha='center',va='center')
ax.text(2,2,0.268,ha='center',va='center')

#Show and save the plot
plt.grid(None)
plt.savefig('40matrix.eps',format='eps', dpi=300) #always put savefig() before plt.show()
plt.show()
#plt.draw()
```

```python
#Matrix plot for a 2D 2*2 system

bacterial = np.array([(0.414, 0.585),(0.357, 0.642)])

#The next two lines are important to initiate the plot
fig = plt.figure()
ax = fig.add_subplot(111)

#Draw the actual plot and colourbar
cax = ax.imshow(bacterial, cmap='hot_r', vmin=0.35, vmax=0.65)
clb = fig.colorbar(cax)
clb.set_label('Fractional Interactions', fontsize='large') #control size of cbar font
#Link for all font-size and customizing features: https://matplotlib.org/api/text_api.html#matplotlib.text.Text

#Put a frame aroun the plot and define its thickness
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams["axes.linewidth"] = 1

#Define the X and Y axes
xnames=['','POPG','','POPE','']
yname=['','','POPE','','','','POPG']
ax.set_xticklabels(xnames, weight='bold', fontsize='x-large')
ax.set_yticklabels(yname, weight='bold', fontsize='x-large')

#Annotate each of the matrix values
ax.text(0,0,0.414,ha='center',va='center', fontsize='large')
ax.text(1,0,0.585,ha='center',va='center', color='w', fontsize='large')
ax.text(0,1,0.357,ha='center',va='center', fontsize='large')
ax.text(1,1,0.642,ha='center',va='center', color='w', fontsize='large')

#Show and save the plot
plt.grid(None)
plt.savefig('bacterial.png',format='png', dpi=300) #always put savefig() before plt.show()
plt.show()
```

```python
#from http://www.learningaboutelectronics.com/Articles/How-to-create-a-matrix-plot-in-seaborn-with-Python.php

import seaborn as sns

%matplotlib inline

flights= sns.load_dataset('flights') #flights is a built-in dataset in seaborn

#print(flights) # prints the dataset
#flights.head() #prints only the initial values in the dataset

fp= flights.pivot_table(index='month', columns='year',values='passengers') #We then convert 'flights' dataset into a matrix using this line

#print(fp)
#sns.heatmap(fp) #displays heatmap in default colours

#some cmap values I liked: Spectral, Blues, Oranges, Greens, summer/summer_r, coolwarm/coolwarm_r
#sns.heatmap(fp, cmap='coolwarm') # displays the heatmap
sns_heatmap = sns.heatmap(fp, cmap='coolwarm')

#following save the plot in high resolution
sns_heatmap.figure.savefig('output_heatmap.eps', format='eps', dpi=400, bbox_inches='tight') #bbox helps save the full figure, otherwise the figure gets chopped off while saving
```
